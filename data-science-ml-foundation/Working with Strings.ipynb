### Pandas String Operations

```python
import pandas as pd
```

```python
monte = pd.Series(['Graham Chapman', 'John Cleese', 'Terry Gilliam',
                   'Eric Idle', 'Terry Jones', 'Michael Palin'])
```

```python
monte
```

```python
monte.str.upper()
```

```python
monte.str.contains('a')
```

```python
monte.str.split()
```

```python
# Read all characters, till you find the condition matching
type(monte.str.extract('([A-Zd-s]+)',expand=False))
```

```python
type(monte.str.extract('([A-Zd-s]+)',expand=True))
```

```python
monte.str.findall(r'^[AEIOU].*[aeiou]$')
```

```python
monte.str[0:3]
```
