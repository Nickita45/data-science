```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

# meshgrid - makes 2 dimensional matrix using 2 1-D array
X,Y = np.meshgrid(x,y)
```

```python
X.shape
```

```python
Y.shape
```

```python
# The pupose is to show some pattern in the data
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# Z is a function of these two matrixs
Z = f(X,Y)
```

```python
Z.shape
```

```python
plt.contour(X,Y,Z, colors='black')
```

```python
plt.contour(X,Y,Z, 20, cmap='RdGy')
plt.colorbar()
plt.savefig('a.png')
```

```python
# The red mean valleys - lesser values
# Back color means rock - Higher values
```
